```python
train_df.isna().sum()
train_df.isnull().any()
train_df=train_df.dropna()
```

```python
train_df.isnull().sum()
```

```python
train_df['srch_ci']=pd.to_datetime(train_df['srch_ci'])
train_df['srch_co']=pd.to_datetime(train_df['srch_co'])
train_df['date_time']=pd.to_datetime(train_df['date_time'])
train_df['Cin_day']=train_df['srch_ci'].apply(lambda a:a.day)
train_df['Cin_month']=train_df['srch_ci'].apply(lambda a:a.month)
train_df['Cin_year']=train_df['srch_ci'].apply(lambda a:a.year)
```

```python
train_df.head()
```

```python
#EDA
sns.displot(train_df['orig_destination_distance'])
plt.title('Destination places close to the origin')
```

```python
ax=train_df['hotel_cluster'].value_counts().plot.bar()
ax.set_xticklabels([])
plt.title("Hotel Clusters")
```

```python
hotel_search=train_df.groupby('hotel_cluster')['srch_destination_id'].nunique()
sns.distplot(hotel_search,bins=50,kde=False)
plt.title("Number of unique search IDs")
```

```python
#Defining X and Y
Y=train_df['is_booking']
X=train_df.drop(['date_time','orig_destination_distance','srch_ci','srch_co','is_booking'],axis=1)
```

```python
X
```

```python
#Train and test data spliting
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
print("Length of X_train ",len(X_train))
print("Length of X_test ",len(X_test))
```

### 1. Logistic Regression

```python
log_reg=LogisticRegression()
log_reg.fit(X_train,Y_train)
y_predict=log_reg.predict(X_test)
accuracy=accuracy_score(Y_test,y_predict)
```

```python
print("Accuracy",accuracy)
print(confusion_matrix(Y_test,y_predict))
print(classification_report(Y_test,y_predict))
```

### 2. KNN Classifier

```python
knn=neighbors.KNeighborsClassifier()
knn.fit(X_train,Y_train)
y_predict=knn.predict(X_test)
accuracy=accuracy_score(Y_test,y_predict)
```

```python
print("Accuracy",accuracy)
print(confusion_matrix(Y_test,y_predict))
print(classification_report(Y_test,y_predict))
```
